```python
#Averaged Data:
import numpy as np
avg = np.loadtxt("annual_csv_average.csv", skiprows=1)
ayr= avg[:,0]
av_year = ayr[::-1]
atp = avg[:,1]
av_temp = atp[::-1]
z = len(av_year)
h=10
davg = np.empty(z, float)

for i in range(z):
    if i==(z-1):
        davg[i]=(av_temp[i] - av_temp[i-1])/h
    else:
        davg[i]=(av_temp[i+1]-av_temp[i-1])/(2*h)


davg[0]=(av_temp[1] - av_temp[0])/h


print("Derivative for Averaged Data:")
print("\n")
print(davg)
print("\n")
print("\n")





#Regular Data:
data = np.loadtxt("annual_csv.csv", skiprows=1)
yr= data[:,0]
year=yr[::-1]
tp= data[:,1]
temp=tp[::-1]
#print(year)
n = len(year)
#print(n)
b=1
dT= np.empty(n, float)


for m in range(n):
    if m==(n-1):
        dT[m]=(temp[m] - temp[m-1])/b
    else:
        dT[m]=(temp[m+1] - temp[m-1])/(2*b)

dT[0]= temp[1]-temp[0]

print("Derivative for Regular Data:")
print("\n")
print(dT)



####################
#Graphing
#File Saving:
import matplotlib.pyplot as plt
#Averages:

fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(av_year, davg)
plt.xlabel('Years')
plt.ylabel('dT/dt')
plt.title('Change in average Earth temperature over time from ten year means')
plt.show()
#fig.savefig("seixas_hw3a.png" ,dpi =300)


#Regular:

fig = plt.figure()                                       
ax = fig.add_subplot(111)
plt.plot(year, dT)                                        
plt.xlabel('Years')                                     
plt.ylabel('dT/dt')                                      
plt.title('Change in average Earth temperature over time')
plt.show()
#fig.savefig("seixas_hw3b.png" ,dpi =300)
```
